# Limitaciones del análisis


1.- No se puede concluir con estos datos la vida del pinguino, porque solo tenemos datos de un año

2.- Las ID no son individuales de cada pinguino, la identificación es general y anual.

3.- ¿la ausencia de registros de algunas especies en determindas islas, se debe a una ausencia real o inaccesibilidad a los nidos?

4.- Datos inexistentes o nulos. La muestra aportada no desvela si es el 100% de la nueva población.





# Conclusiones:

## En cuanto a datos:

1.- El dataset no es equilibrado en cuanto a sus especies. No todas las especies de pinguinos están en todas las islas

2.- El dataset tiene falta de datos. Principalmente en el sexo de los pinguinos

3.- Detectamos tres variables: especies, islas y sexo

4.-La mitad de los pingüinos se encuentra en la isla Biscoe, una minoria esta en la isla Torgersen, y los demás en la isla Dream. 

5.- Solo se registra el primer año de vida del pinguino en cada estudio.

## En cuanto a morfologia:

1.- Los pingüinos Gentoo son los más faciles de distinguir:
    Son los mas grandes en los dos sexos.
    Tienen una profundidad de pico menor. 
    Cuentan con una longitud de aleta y una masa corporal superior a las demás.

2.- Las especies Adelie y Chinstrap suelen tener el mismo tamaño y características mezcladas como, longitud de aleta, peso etc.. A veces son dificiles de distinguir.

3.- Los pingüinos Adelie generalmente tienen una longitud de pico menor.
