## 3. Generar un Data frame desde la base de datos.csv

```python
import pandas as pd
df = pd.read_csv ("data/penguins_raw.csv")
df

```

## 4 Consultas bÃ¡sicas, nÃºmero de individuos, especies y distribucion en las islas

```python
# Contar cuantas especies se han detectado
conteo_de_espeies = df.groupby('Species').size().reset_index(name='Count')
print("Las especies detectadas son: ")
print(conteo_de_espeies, "\n")
df


# Contar cuantas especies se han detectado por isla
avistamientos_por_isla = df.groupby(['Species', 'Island']).size().reset_index(name='Count')
print("Los avistamientos de especies por isla son: ")
print(avistamientos_por_isla, "\n")

df['Island'].value_counts()

# Contar en cuÃ¡ntas islas aparece cada individuo
islands_por_individuo = df.groupby('Individual ID')['Island'].nunique()

# Filtrar los que aparecen en mÃ¡s de una isla
duplicados = islands_por_individuo[islands_por_individuo > 1]

if not duplicados.empty:
    print("Individual IDs presentes en mÃºltiples islas:")
else:
    print("No hay Individual IDs en mÃ¡s de una isla.")

detalle = df.groupby('Individual ID')['Island'].unique()

duplicados_detalle = detalle[detalle.apply(len) > 1]

print(duplicados_detalle)
```

## 5 CorreciÃ³n de formato de fecha

```python
def parse_fecha(x):
    for fmt in ('%Y-%m-%d', '%d/%m/%Y', '%m/%d/%Y', '%Y/%m/%d'):
        try:
            return pd.to_datetime(x, format=fmt)
        except:
            continue
    return pd.NaT  # si ninguno funciona

df['Date Egg'] = df['Date Egg'].apply(parse_fecha)

# Ahora sÃ­ formateas
df['Date Egg'] = df['Date Egg'].dt.strftime('%Y/%m/%d')
df
```


# 6 Comprobar que la columna Stage y Region siempre tienen el mismo dato y suprimirla del DF


```python
#comprobar que  el dato se repite en todas las filas
for col in ['Stage', 'Region']:
    conteo = df[col].value_counts(dropna=False)
    
    print(f"\nDatos en la columna {col}:")
    print(conteo)
    
    if df[col].nunique(dropna=False) == 1:
        print(f"â†’ Eliminando '{col}' (solo tiene un valor)")
        df = df.drop(columns=[col])
    else:
        print(f"â†’ '{col}' se mantiene (tiene mÃºltiples valores)")

df
```

## 7 Suprimir la palabra "Penguin" del string que aparece en la column Species

```python

df['Species'] = df['Species'].apply(lambda x: x.replace('Penguin', '') if type(x) == str else x)
df

```

## 8 Convertir todo a minusculas

```python
# ðŸ”¤ Columnas en minÃºsculas
df.columns = df.columns.str.lower()

# ðŸ”¤ Valores en minÃºsculas (robusto)
for col in df.select_dtypes(include=["object", "string"]):
    df[col] = df[col].str.lower()

df
```

## 9 Df limpio sin filas en donde haya 4 valores NaN en diferentes columns

```python

df_clean = df.dropna(thresh=df.shape[1] - 3)
df_clean
```

## 10 simplificacion de texto y caracteres

```python

import re
df_clean['species'] = df_clean['species'].str.replace(
    r'\([^)]*\)', '', regex=True
)
df_clean['species'] = df_clean['species'].str.replace('penguin', '', regex=False)


df_clean.columns = [col.replace(' ', '_') for col in df_clean.columns]

df_clean
```

## 11 Eliminar columnas que no aportan datos relevantes al estudio

```python

df_clean = df_clean.drop(columns=['delta_15_n_(o/oo)', 'delta_13_c_(o/oo)', 'comments'])
df_clean
```

## 12 borrador CSV

```python

df_clean.to_csv('data/df_clean_penguins.csv', index=False)
```

## 13 Nuevo conteo de nÃºmero de individuos, especies y distribucion en las islas con el Df limpio

```python
# Contar cuantas especies se han detectado
conteo_de_espeies = df_clean.groupby('species').size().reset_index(name='Count')
print("Las especies detectadas son: ")
print(conteo_de_espeies, "\n")
df


# Contar cuantas especies se han detectado por isla
avistamientos_por_isla = df_clean.groupby(['species', 'island']).size().reset_index(name='Count')
print("Los avistamientos de especies por isla son: ")
print(avistamientos_por_isla, "\n")

df_clean['island'].value_counts()

# Contar en cuÃ¡ntas islas aparece cada individuo
islands_por_individuo = df_clean.groupby('individual_id')['island'].nunique()

# Filtrar los que aparecen en mÃ¡s de una isla
duplicados = islands_por_individuo[islands_por_individuo > 1]

if not duplicados.empty:
    print("individual ids presentes en mÃºltiples islas:")
else:
    print("No hay Individual IDs en mÃ¡s de una isla.")

detalle = df_clean.groupby('individual_id')['island'].unique()

duplicados_detalle = detalle[detalle.apply(len) > 1]

print(duplicados_detalle)
```
