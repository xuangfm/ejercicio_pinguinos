## 3. Generar un Data frame desde la base de datos.csv

```python
import pandas as pd
df = pd.read_csv ("penguins_raw.csv")
df

```

## 4 Consultas básicas, número de individuos, especies y distribucion en las islas

```python
# Contar cuantas especies se han detectado
conteo_de_espeies = df.groupby('Species').size().reset_index(name='Count')
print("Las especies detectadas son: ")
print(conteo_de_espeies, "\n")
df


# Contar cuantas especies se han detectado por isla
avistamientos_por_isla = df.groupby(['Species', 'Island']).size().reset_index(name='Count')
print("Los avistamientos de especies por isla son: ")
print(avistamientos_por_isla, "\n")

df['Island'].value_counts()

# Contar en cuántas islas aparece cada individuo
islands_por_individuo = df.groupby('Individual ID')['Island'].nunique()

# Filtrar los que aparecen en más de una isla
duplicados = islands_por_individuo[islands_por_individuo > 1]

if not duplicados.empty:
    print("Individual IDs presentes en múltiples islas:")
else:
    print("No hay Individual IDs en más de una isla.")

detalle = df.groupby('Individual ID')['Island'].unique()

duplicados_detalle = detalle[detalle.apply(len) > 1]

print(duplicados_detalle)
```

## 5 Correción de formato de fecha

```python
def parse_fecha(x):
    for fmt in ('%Y-%m-%d', '%d/%m/%Y', '%m/%d/%Y', '%Y/%m/%d'):
        try:
            return pd.to_datetime(x, format=fmt)
        except:
            continue
    return pd.NaT  # si ninguno funciona

df['Date Egg'] = df['Date Egg'].apply(parse_fecha)

# Ahora sí formateas
df['Date Egg'] = df['Date Egg'].dt.strftime('%Y/%m/%d')
df
```


# 6 Comprobar que la columna Stage y Region siempre tienen el mismo dato y suprimirla del DF


```python
#comprobar que  el dato se repite en todas las filas
for col in ['Stage', 'Region']:
    conteo = df[col].value_counts(dropna=False)
    
    print(f"\nDatos en la columna {col}:")
    print(conteo)
    
    if df[col].nunique(dropna=False) == 1:
        print(f"→ Eliminando '{col}' (solo tiene un valor)")
        df = df.drop(columns=[col])
    else:
        print(f"→ '{col}' se mantiene (tiene múltiples valores)")

df
```

## 7 Suprimir la palabra "Penguin" del string que aparece en la column Species

```python

df['Species'] = df['Species'].apply(lambda x: x.replace('Penguin', '') if type(x) == str else x)
df
```

```python
# 🔤 Columnas en minúsculas
df.columns = df.columns.str.lower()

# 🔤 Valores en minúsculas (robusto)
for col in df.select_dtypes(include=["object", "string"]):
    df[col] = df[col].str.lower()

df
```

## Df limpio sin filas en donde haya 4 valores NaN en diferentes columns

```python

df_clean = df.dropna(thresh=df.shape[1] - 3)
df_clean
```

## Eliminar columnas que no aportan datos relevantes al estudio

```python

df_clean = df_clean.drop(columns=['delta 15 n (o/oo)', 'delta 13 c (o/oo)', 'comments'])
df_clean
```

## 1er borrador CSV

```python

df_clean.to_csv('df_clean_penguins.csv', index=False)
```
